```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/file.tsv', delimiter= '\t')
df.head()
```

```python
df.groupby('Country')['Age'].mean()    # we are working on Country column, which we seperated, 
# then we took the mean of the ages for each country values.


# it's breaking up values in Country, we can use each value of country seperately as well.
```

```python
# now if we want to take data for only Algeria then we use loc,

Coun_Algeria = df.loc[df['Country'] == 'Algeria', :]
Coun_Algeria
```

```python
# Now we'll take the age dataset of algeria,

Coun_Algeria['Age']
```

```python
# mean for algeria age dataset,

Coun_Algeria['Age'].mean()
```

```python
# lets use describe().

df.groupby('Country')['Selections'].describe()    # std stands for standard deviation.
```

```python


# We can write our 0wn functions or numpy fuctions for groupby as well. For that we have to use " .agg 
# or .aggregate(np.function_name)"  in case for numpy, e.g
 

j = df.groupby('Country')['Age']   
j.agg(np.mean)
# gives same result
```

```python

# writing mean() function

def my_func(values):
    n = len(values)
    s = np.sum(values)
    return s / n

# now use j object that we made.
j.agg(my_func)
# gives same result as well.
```

## In above function of mean( ) we made, it was a vectorized function, because we used numpy sum method in it for making it faster instead of using for loops.

## By usingg .agg or .aggregate , we can pass multiple functions, and we can also pass multiple things as well, e.g

### When writing multiple functions, make sure to write the command from the beginning and not from a saved variable cause that sometimes don't work.

```python
df.groupby('Country')['Age'].agg([
    np.mean,
    np.std,
    np.count_nonzero
])
```

```python

# same example but we'll run the function on multiple things ,

df.groupby(['Country', 'Club', 'Group'])['Age'].agg([
    np.mean,
    np.std,
    np.count_nonzero
]).reset_index()    # used reset index for a suitable format.
```

```python

#  WE CAN USE DICTIONARIES TO RE-NAME COLUMNS,
df.groupby('Country')['Age'].agg({
    'MY-mean' : np.mean,
    'Standard deviation' : np.std,
    'Non-Zero_Values' : np.count_nonzero
})

# this is not an error , it's a warning because it's old method of doing things.
```

```python


# NEW METHOD OF RE-NAMING COLUMNS,

df.groupby('Country')['Age'].agg([
    np.mean,
    np.std,
    np.count_nonzero
]).rename(columns = {'count_nonzero' : 'NonZero', 'mean' : 'avg', 'std' : 'Standard Deviation'}).reset_index()

```
